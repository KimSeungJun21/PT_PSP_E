```python
import torch

# 확인하고 싶은 데이터 파일 경로 (예시)
data_path = "/home/kimseungjun/datasets/S3DIS_preprocess/Area_3/hallway_2.pth"

try:
    # 데이터 로드
    data = torch.load(data_path)
    
    print(f"--- 파일 정보: {data_path} ---")
    
    # 1. 존재하는 모든 Key 확인
    print(f"데이터에 포함된 항목(Keys): {data.keys()}")
    
    # 2. 'normal' 항목 존재 여부 확인
    if 'normal' in data:
        print("\n✅ [결과] 'normal' 데이터가 존재합니다!")
        print(f"'normal' 데이터의 형태(Shape): {data['normal'].shape}")
    else:
        print("\n❌ [결과] 'normal' 데이터가 없습니다.")
        print("사용 가능한 항목만 나열:", list(data.keys()))

except Exception as e:
    print(f"파일을 읽는 중 에러 발생: {e}")
```
